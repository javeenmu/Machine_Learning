```python
# Importing libraries
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating Dataset (650 instances)
rd.seed(700)
data = []

# 650 instances using for looop
for _ in range(650):
    height= round(rd.uniform(4.5,7.0), 2)
    weight = rd.randint(50000, 100000)
    data.append((height, weight))

# Printing the dataset
print("Raw Dataset")
print("Height (ft) Weight (gms)")
for record in data:
    print(f"{record[0]:<11} {record[1]:<11}")

weights = np.array([record[1] for record in data])
```

```python
#Normalization using Min-Max
min_w = np.min(weights)
max_w = np.max(weights)
minmax_normalized_weight = (weights - min_w)/(max_w - min_w)

#Display Normalized weights
print("Normalized Weights using Min-Max Algorithm")
for i, weight in enumerate(minmax_normalized_weight):
    print(f"Record {i+1}: {weight: .4f}")
```

```python
# Normalization using Z-score
mean_w = np.mean(weights)
stddev_w = np.std(weights)
zscore_normalized_weight = (weights - mean_w)/stddev_w

#Display Normalized weights
print("Normalized Weights using Z-Score")
for i, weight in enumerate(zscore_normalized_weight):
    print(f"Record {i+1}: {weight: .4f}")
```
