```python
#Importing libraries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
#Loading Dataset
data = pd.read_csv("WisconsinData.csv")
data.drop(columns=['id'], inplace=True)
data.drop(columns=['Unnamed: 32'], inplace=True)
data['diagnosis'] = data['diagnosis'].map({'B': 0, 'M': 1})
data.head(5)
```

```python
#Divide into train and test
Y = data['diagnosis']
X = data.drop('diagnosis', axis=1) 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Implement Decision tree classifier
classifier = DecisionTreeClassifier(criterion='gini')
classifier.fit(X_train, Y_train)
Y_pred = classifier.predict(X_test)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
# Performance Metrics
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
accuracy = accuracy_score(Y_test, Y_pred)

# Sensitivity (Recall for Positive Class) and Specificity (Recall for Negative Class)
sensitivity = recall  # Sensitivity is the same as recall for the positive class
specificity = cm[0, 0] / (cm[0, 0] + cm[0, 1])  # True Negative Rate

print(f"Precision: {precision:.4f}")
print(f"Recall (Sensitivity): {recall:.4f}")
print(f"Specificity: {specificity:.4f}")
print(f"Accuracy: {accuracy:.4f}")

# Classification report for more details
print("\nClassification Report:")
print(classification_report(Y_test, Y_pred, target_names=['Benign', 'Malignant']))
```
